```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList():
    def __init__(self):
        self.head = None
        
    
    def insert_at_end(self, entry):
        newnode = Node(entry)
        if self.head == None:
            self.head = newnode
        else:
            lastnode = self.head
            while True:
                if lastnode.next == None:
                    break
                else:
                    lastnode = lastnode.next
            lastnode.next = newnode
        
    
    def insert_at_beginning(self, entry):
        newnode = Node(entry)
        newnode.next = self.head
        self.head = newnode
    
    
    def insert_at(self, entry, index):
        if index == 0:
            self.insert_at_beginning(entry)
            return
        
        if index < 0 or index > self.get_length():
            raise Exception('Invalid Index')
            return
            
        newnode = Node(entry)
        count = 0
        currentnode = self.head        
        
        while currentnode:            
            if count == index - 1 :
                newnode.next = currentnode.next
                currentnode.next = newnode
                break
            currentnode = currentnode.next
            count += 1
    
    
    def insert_after_value(self, value_to_search, entry):
        newnode = Node(entry)
        currentnode = self.head
        found = False
        
        while True:
            if currentnode == None:
                break
            else:
                if currentnode.data == value_to_search:
                    newnode.next = currentnode.next
                    currentnode.next = newnode
                    found = True
                    break
                else:
                    currentnode = currentnode.next
        
        if found == False:
            raise Exception('Number not found')

    
    def remove_at(self, index):        
        count = 0
        currentnode = self.head
         
        if index == 0:
            self.head = self.head.next
            self.print_list()
            return

        while currentnode:
            if count == index - 1:
                currentnode.next = currentnode.next.next
                break
            currentnode = currentnode.next
            count += 1
            
        self.print_list()
    
    
    
    def remove_by_value(self, value_to_remove):
        currentnode = self.head
        found = False
        
        if value_to_remove == self.head.data:
            self.head = self.head.next
            self.print_list()
            return
        
        while True:
            if currentnode == None:
                break
            else:            
                if currentnode.data == value_to_remove:
                    tempnode.next = currentnode.next
                    found = True
                    self.print_list()
                    break
                else:
                    tempnode = currentnode
                    currentnode = currentnode.next
        
        if found == False:
            raise Exception('Number nor found')
    
    
    def get_length(self):
        count = 0
        lastnode = self.head
        while lastnode.next:
            count += 1
            lastnode = lastnode.next
        length = count + 1
        return length
        
        
    def print_list(self):
        if self.head == None:
            print('List is empty')
            return
        
        currentnode = self.head
        llstr = ''
        while currentnode:
            llstr += (str(currentnode.data) +  " --> ") if currentnode.next else str(currentnode.data)
            currentnode = currentnode.next
        print(llstr)
        
    
    def insert_values(self, entry):
        for e in entry:
            self.insert_at_end(e)
            
            
if __name__ == '__main__':
    ll = LinkedList()
    ll.insert_values(['2', '3', '4'])
    ll.insert_at_end('5')
    ll.insert_at_beginning('1')
    ll.insert_at('a', 2)

    ll.print_list()

    ll.remove_at(1)
```
