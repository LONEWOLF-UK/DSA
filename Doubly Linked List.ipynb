```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class LinkedList():
    def __init__(self):
        self.head = None
        
    
    def insert_at_end(self, entry):
        newnode = Node(entry)
        if self.head == None:
            self.head = newnode
        else:
            currentnode = self.head
            while True:
                if currentnode.next == None:
                    break
                else:
                    currentnode = currentnode.next
            newnode.prev = currentnode
            currentnode.next = newnode
        
    
    def insert_at_beginning(self, entry):
        newnode = Node(entry)
        newnode.next = self.head
        self.head.prev = newnode
        self.head = newnode
    
    
    def insert_at(self, entry, index):
        if index == 0:
            self.insert_at_beginning(entry)
            return
        
        if index < 0 or index >= self.get_length():
            raise Exception('Invalid Index')
            return
        
        if index == self.get_length() - 1:
            self.insert_at_end(entry)
            return
            
        newnode = Node(entry)
        count = 0
        currentnode = self.head        
        
        while currentnode:
            if count == index - 1 :
                newnode.next = currentnode.next
                currentnode.next.prev = newnode
                newnode.prev = currentnode
                currentnode.next = newnode
                break
            currentnode = currentnode.next
            count += 1
    
    
    def insert_after_value(self, value_to_search, entry):
        newnode = Node(entry)
        currentnode = self.head
        found = False
        
        while True:
            if currentnode == None:
                break
            else:
                if currentnode.data == value_to_search and currentnode.next == None:
                    newnode.prev = currentnode
                    currentnode.next = newnode
                    newnode.next = None
                    return
                
                if currentnode.data == value_to_search:
                    newnode.next = currentnode.next
                    newnode.prev = currentnode
                    currentnode.next.prev = newnode
                    currentnode.next = newnode
                    found = True
                    break
                else:
                    currentnode = currentnode.next
        
        if found == False:
            raise Exception('Number not found')

    
    def remove_at(self, index): 
        if index < 0 or index >= self.get_length():
            raise Exception('Invalid Index')
            return
        
        count = 0
        currentnode = self.head
         
        if index == 0:
            self.head = self.head.next
            self.head.prev = None
            self.print_forward()
            self.print_backward()
            return
            
+
        while currentnode:
            if currentnode.next.next == None:
                currentnode.next = None
                self.print_forward()
                self.print_backward()
                return
            if count == index - 1:
                currentnode.next.next.prev = currentnode.next.prev
                currentnode.next = currentnode.next.next
                break
            currentnode = currentnode.next
            count += 1
    
        self.print_forward()
        self.print_backward()
        
    
    def remove_by_value(self, value_to_remove):
        currentnode = self.head
        found = False

        if value_to_remove == self.head.data:
            self.head = self.head.next
            if self.head:
                self.head.prev = None
            self.print_forward()
            self.print_backward()
            return

        while currentnode:
            if currentnode.data == value_to_remove:
                if currentnode.next:
                    currentnode.next.prev = currentnode.prev
                currentnode.prev.next = currentnode.next
                del currentnode
                found = True
                break
            currentnode = currentnode.next

        if not found:
            raise Exception('Number not found')

        self.print_forward()
        self.print_backward()
    
    
    
    def get_length(self):
        length = 0
        currentnode = self.head
        while currentnode:
            length += 1
            currentnode = currentnode.next
        return length
        
        
        
    def print_forward(self):
        if self.head == None:
            print('List is empty')
            return
        
        currentnode = self.head
        llstr = ''
        while currentnode:
            llstr += (str(currentnode.data) +  " --> ") if currentnode.next else str(currentnode.data)
            currentnode = currentnode.next
        print(llstr)
    
    
    
    def print_backward(self):
        if self.head == None:
            print('List is empty')
            return
        
        currentnode = self.head
        llstr = ''
        while True:
            if currentnode.next == None:
                break
            else:
                currentnode = currentnode.next
        
        while currentnode:
            llstr += str(currentnode.data + " <-- ") if currentnode.prev else str(currentnode.data)
            currentnode = currentnode.prev
        print(llstr)
    
    
    
    def insert_values(self, entry):
        for e in entry:
            self.insert_at_end(e)
            
            
if __name__ == '__main__':
    ll = LinkedList()
    ll.insert_values(['2', '3', '4'])
    ll.insert_at_end('5')
    ll.insert_at_beginning('1')
    
    ll.insert_at('a', 1)
    
    ll.remove_at(5)
```
